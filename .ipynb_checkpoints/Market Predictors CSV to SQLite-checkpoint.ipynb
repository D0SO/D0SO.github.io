```python
# Dependencies
# ----------------------------------
# Imports the method used for connecting to DBs
from sqlalchemy import create_engine

# Imports the methods needed to abstract classes into tables
from sqlalchemy.ext.declarative import declarative_base

# Allow us to declare column types
from sqlalchemy import Column, Integer, String, Float 

import pandas as pd
```

```python
path_to_bonds_value_csv="Data Treatment/Bonds (closing price).csv"
bonds_df=pd.read_csv(path_to_bonds_value_csv)

path_to_bonds_value_csv="Data Treatment/Bonds (daily returns).csv"
bondsdaily_df=pd.read_csv(path_to_bonds_value_csv)

path_to_bonds_value_csv="Data Treatment/Gold (closing price).csv"
gold_df=pd.read_csv(path_to_bonds_value_csv)

path_to_bonds_value_csv="Data Treatment/Nasdaq (closing price).csv"
nasdaq_df=pd.read_csv(path_to_bonds_value_csv)

path_to_bonds_value_csv="Data Treatment/Nasdaq (daily returns).csv"
nasdaqdaily_df=pd.read_csv(path_to_bonds_value_csv)

path_to_bonds_value_csv="Data Treatment/Nikkei (closing price).csv"
nikkei_df=pd.read_csv(path_to_bonds_value_csv)

path_to_bonds_value_csv="Data Treatment/Nikkei (daily returns).csv"
nikkeidaily_df=pd.read_csv(path_to_bonds_value_csv)
```

```python
# Create Database Connection
# ----------------------------------
# Creates a connection to our DB
# Do not duplicate this cell. There is no need 
engine = create_engine("sqlite:///market.sqlite")
conn = engine.connect()
```

```python
bonds_df.to_sql("BondsClosing",engine,if_exists="replace")

bondsdaily_df.to_sql("BondsDaily",engine,if_exists="replace")

gold_df.to_sql("GoldClosing",engine,if_exists="replace")

nasdaq_df.to_sql("NasdaqClosing",engine,if_exists="replace")

nasdaqdaily_df.to_sql("NasdaqDaily",engine,if_exists="replace")

nikkei_df.to_sql("NikkeiClosing",engine,if_exists="replace")

nikkeidaily_df.to_sql("NikkeiDaily",engine,if_exists="replace")
```
