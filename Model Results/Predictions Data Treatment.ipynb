```python
import pandas as pd
import numpy as np
```

```python
# Date time information
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html
```

```python
nasdaq_r = "Nasdaq-Returns-Predicted.csv"
nasdaq_v = "Nasdaq-Volatility-Predicted.csv"
nikkei_r = "Nikkei-Returns-Predicted.csv"
nikkei_v = "Nikkei-Volatility-Predicted.csv"
gold_r = "Gold-Returns-Predicted.csv"
gold_v = "Gold-Volatility-Predicted.csv"
bond_r = "10y-Bonds-Return-Predicted.csv"
bond_v = "10y-Bonds-Volatility-Predicted.csv"

nasdaq_r = pd.read_csv(nasdaq_r)
nasdaq_v = pd.read_csv(nasdaq_v)
nikkei_r = pd.read_csv(nikkei_r)
nikkei_v = pd.read_csv(nikkei_v)
gold_r = pd.read_csv(gold_r)
gold_v = pd.read_csv(gold_v)
bond_r = pd.read_csv(bond_r)
bond_v = pd.read_csv(bond_v)
```

```python
predictions = pd.DataFrame ({
        # 'count of days' : np.arange(len(gold_r)),
        'nasdaq_actual_return' : nasdaq_r['actual_returns'],
        'nasdaq_predicted_return' : nasdaq_r['predicted_returns'],
        'nasdaq_actual_volatility' : nasdaq_v['actual_volatility'],
        'nasdaq_predicted_volatility': nasdaq_v['predicted_volatility'],
        'nikkei_actual_return' :  nikkei_r['actual_returns'],
        'nikkei_predicted_return' : nikkei_r['predicted_returns'],
        'nikkei_actual_volatility' : nikkei_v['actual_volatility'],
        'nikkei_predicted_volatility': nikkei_v['predicted_volatility'],
        'gold_actual_return' : gold_r['actual_returns'],
        'gold_predicted_return' : gold_r['predicted_returns'],
        'gold_actual_volatility' : gold_v['actual_volatility'],
        'gold_predicted_volatility' : gold_v['predicted_volatility'],
        '10y_bond_actual_return' : bond_r['actual_returns'],
        '10y_bond_predicted_return' : bond_r['predicted_returns'],
        '10y_bond_actual_volatility' : bond_v['actual_volatility'],
        '10y_bond_predicted_volatility' : bond_v['predicted_volatility']
        
})

predictions.head()
```

```python

returns_actual = predictions[['nasdaq_actual_return','nikkei_actual_return','gold_actual_return','10y_bond_actual_return']]

returns_prediction = predictions[['nasdaq_predicted_return','nikkei_predicted_return','gold_predicted_return','10y_bond_predicted_return']]

volatility_actual = predictions[['nasdaq_actual_volatility','nikkei_actual_volatility','gold_actual_volatility','10y_bond_actual_volatility']]

volatility_prediction = predictions[['nasdaq_predicted_volatility','nikkei_predicted_volatility','gold_predicted_volatility','10y_bond_predicted_volatility']]

```

```python
#Create mean column
returns_actual['averager_return'] = returns_prediction.mean(axis=1)
returns_prediction['average_return'] = returns_prediction.mean(axis=1)
volatility_actual['average_volatility'] = volatility_actual.mean(axis=1)
volatility_prediction['average_volatility'] = volatility_prediction.mean(axis=1)

returns_actual['day_count'] = np.arange(len(returns_actual))
returns_prediction['day_count'] = np.arange(len(returns_prediction))
volatility_actual['day_count'] = np.arange(len(volatility_actual))
volatility_prediction['day_count'] = np.arange(len(volatility_prediction))
```

```python
#  Drop N/a's 
returns_actual = returns_actual.dropna()
returns_prediction = returns_prediction.dropna()
volatility_actual = volatility_actual.dropna()
volatility_prediction = volatility_prediction.dropna()
```
